```python
df_header = pd.read_excel('header_img.xlsx',sheet_name='IMG')
df_header_lbl = pd.read_excel('header lbl.xlsx',sheet_name='Feuille 1')
```

```python
df_data.columns=columns
df_data_lbl.columns = columns_lbl
```
